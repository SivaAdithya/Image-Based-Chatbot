```python
import humingbird
import random
import cv2
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
import numpy as np
from googlesearch import search
```

```python
color_model = keras.models.load_model('color1.h5')
```

```python
colors = ['black', 'white', 'red', 'blue', 'green', 'yellow', 'purple', 'gray', 'brown', 'pink']
```

```python
def find_color(path):
    img = cv2.imread(path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    img = cv2.resize(img,(28,28))
    img =img.reshape(1,28,28,3)
    x = color_model.predict(img)
    ret = np.argmax(x, axis=1)
    return colors[ret[0]]
```

```python
clothes = ["sweatshirt","formal shirt","t-shirt","frock","jean","skirt","formal pant","sweat pant",
           "saree","kurti","jacket","hoodie","blazer","suit","dhoti","salwar","shorts","jumpsuit",
           "playsuit","lehenga","ghagra","sharara","palazzo pant"]

pretext = ["The image you have uploaded is a ","This is an image of ","This seem like a ",
          "In my opinion this is a ","This must be a ","My CPU tells me this is a ","Based on the result of my threads being executed I can tell this is a "]
intents = {}
ct = 0
for i in clothes:
    intents[i] = ct
    ct = ct+1
visual_intents = {}
for i in clothes:
    visual_intents[i] = random.choice(pretext) + i

print(intents)
print(visual_intents)
```

```python
def detect_and_respond(query):
    prediction = humingbird.Text.predict(
    text=query,
    labels=["sweatshirt","formal shirt","t-shirt","frock","jean","skirt","formal pant","sweat pant",
           "saree","kurti","jacket","hoodie","blazer","suit","dhoti","salwar","shorts","jumpsuit",
           "playsuit","lehenga","ghagra","sharara","palazzo pant"]
  )
  
    highest_score = 0
    highest_score_class = ""
  
    for i in prediction:
        if i["score"] > highest_score:
            highest_score = i["score"]
            highest_score_class = i["className"]
    return intents[highest_score_class]


def visual_intent_detection(image):
    prediction = humingbird.Image.predict(
    image_path=image,
    labels=["sweatshirt","formal shirt","t-shirt","frock","jean","skirt","formal pant","sweat pant",
           "saree","kurti","jacket","hoodie","blazer","suit","dhoti","salwar","shorts","jumpsuit",
           "playsuit","lehenga","ghagra","sharara","palazzo pant"]
  )
  
    highest_score = 0
    highest_score_class = ""
    for i in prediction:
        if i["score"] > highest_score:
            highest_score = i["score"]
            highest_score_class = i["className"]
    return (visual_intents[highest_score_class],intents[highest_score_class],highest_score_class)
```

```python
texts = ["Using the following link find our latest collections of ","Using the link below check out our collections of "
        ,"Click the link below to find a wide varities of available ", "Flat 20% off on all purchases!! Click the link to view "]
dress = {0:'sweatshirt', 1:'formal shirt', 2: 't-shirt', 3:'frock', 4:'jean',
         5:'skirt', 6:'formal pant', 7:'sweat pant', 8:'saree', 9:'kurti', 10:'jacket', 11:'hoodie', 12:'blazer',13:'suit',
         14:'dhoti', 15:'salwar', 16:'shorts', 17:'jumpsuit', 18:'playsuit', 19:'lehenga', 20:'ghagra', 21:'sharara',
         22:'palazzo pant'}
link = {0:"",
       1: " https://www.myntra.com/shirt",
       2: " https://www.myntra.com/tshirt",
       3: " https://www.myntra.com/frock",
       4: " https://www.myntra.com/jeans",
       5: " https://www.myntra.com/skirt",
       6: " https://www.myntra.com/pants",
       7: " https://www.myntra.com/cargo",
       8: " https://www.myntra.com/saree",
       9: " https://www.myntra.com/kurti",
       10: " https://www.myntra.com/jacket",
       11: " https://www.myntra.com/hoodie",
       12: " https://www.myntra.com/blazer",
       13: " https://www.myntra.com/suit",
       14: " https://www.myntra.com/dhoti",
       15: " https://www.myntra.com/salwar",
       16: " https://www.myntra.com/cargo",
       17: " https://www.myntra.com/jumpsuit",
       18: " https://www.myntra.com/playsuit",
       19: " https://www.myntra.com/lehenga",
       20: " https://www.myntra.com/lehenga",
       21: " ###",
       22: " ###"}
def print_message(val):
    msg1 = random.choice(texts)
    msg2= " " + dress[val] + " ,link:" + link[val]
    return msg1+msg2
```

```python
greet = ['Hey!!How may I be of assistance?','Hello mate!! Looking good today! How may I help you look better?',
        'Greetings breathable being! I can help make your shopping experience fun and simple']

bye = ['Glad I could be of assistance! Be sure to hit me up if you need anything else',
      'Bye bye! I will be here to provide further assistance if needed',
      'Hope I was able to satisfy all your needs!']
invalid = ["Sorry, I am not good at maths. Please enter a number between 1 to 4 as those are the only numbers I know",
          "My creators did not teach me numbers beyond 4 as I think they lack the knowledge themselves. Hence at no fault of mine I am unable to understand your input",
          "Given that I am a cognitive agent I should know numbers beyond 4. But I do not, so please enter any of the first 4 natural numbers",
          "I am not as smart as you to be able to understand numbers beyond 4. Hence kindly tone down your smartness and enter values from 1 to 4"
          ]
```

```python

while(True):
    print(random.choice(greet))
    print("Press 1 to type your question")
    print("Press 2 to enter the link of the image you want to upload")
    print("Press 3 to be greeted again")
    print("Press 4 to exit")
    val = int(input())
    if(val==1):
        query = input("Input your query")
        print(detect_and_respond(query))
    elif(val==2):
        img = input("Paste the path of the image here")
        res = visual_intent_detection(img)
        print(res[0])
        print(print_message(res[1]))
        val = input("Do you want similar color recomendations? y/n")
        if(val=='y' or val =='Y'):
            c = find_color(img)
            print("Here are some more",c,res[2])
            query = c+ ' ' + res[2]
            for j in search(query, tld="co.in", num=10, stop=10, pause=2):
                print(j)
            #print(link[c])
    elif(val==3):
        continue
    elif(val==4):
        print(random.choice(bye))
        break
    else:
        print(random.choice(invalid))
        print("\n\n")
    print("\n\n")
```
